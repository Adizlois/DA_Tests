```python

#%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (15.0, 7.50)
from scipy.stats import norm
import numpy as np
def plot_dist(mean,sd,col,name):
    x = np.linspace(mean-5*sd,mean+5*sd,5000)
    y = norm.pdf(x, mean, sd) 
    plt.plot(x,y,color=col,label=name)
    
def multiply(mean1,mean2,sd1,sd2):
    x=np.linspace(-10,10,1000)
    y1=norm.pdf(x,mean1,sd1)
    y2=norm.pdf(x,mean2,sd2)
    y3=y1*y2
    xmaximum=x[np.where(y3==y3.max())]

    print "Maximum (multiplication)= ",xmaximum
    plt.plot(x,y3,color="yellow",label="Multiplication")
    
def DA_univariate_gaussian(meanprior,sdprior,meanlikelihood,sdlikelihood):
    mean_posterior=meanprior+(sdprior/(sdprior+sdlikelihood))*(meanlikelihood-meanprior)
    sd_posterior=sdprior*(1-(sdprior/(sdprior+sdlikelihood)))
    print "Mean posterior = ", mean_posterior
    print "Sd Posterior = ", sd_posterior
    a=plot_dist(mean_posterior,sd_posterior,"blue","Posterior")
    b=plot_dist(meanprior,sdprior,"red","Prior")
    c=plot_dist(meanlikelihood,sdlikelihood,"black","Likelihood")
    d=multiply(meanprior,meanlikelihood,sdprior,sdlikelihood)
    plt.legend()
    
    
DA_univariate_gaussian(-5.,1.,0.,4.)
plt.show()
```
